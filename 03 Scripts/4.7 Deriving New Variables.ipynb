```python
import pandas as pd
import os



path = r'C:\\Users\\mutta\\Documents\\Project 4\\' 


ords_prods_merged_pkl = os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl')

df_ords_prods_merged  = pd.read_pickle(ords_prods_merged_pkl)

df_ords_prods_merged.head(10)
```

```python
pd.options.display.float_format = '{:.2f}'.format

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```




### create the “price_label” and “busiest_day” columns

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_label'] = 'High-range product'
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5),
                                                 'price_label'] = 'Mid-range product' 
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_label'] = 'Low-range product'
```

```python
result = []

print(color.BLUE + color.BOLD + 'Creating busiest day list')

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
    
df_ords_prods_merged['busiest_day'] = result

print(df_ords_prods_merged.columns)
```

```python

print(color.BLUE + color.BOLD +'Total number of orders in a day of the week' + color.END)
print(df_ords_prods_merged["orders_day_of_week"].value_counts(dropna=False))

print('-----------------------------------------------------------')

print(color.BLUE + color.BOLD + 'Classification of busy days of the week'+ color.END)
print(df_ords_prods_merged['busiest_day'].value_counts(dropna=False))


print('-----------------------------------------------------------')

print(color.BLUE + color.BOLD +'Classification of price orders'+ color.END)
print(df_ords_prods_merged['price_label'].value_counts(dropna=False))

print('-----------------------------------------------------------')
```

### Reclassification of busy days of the week

```python
df_ords_prods_merged.loc[df_ords_prods_merged['orders_day_of_week'] <= 1, 'busy_days_of_the_week'] = 'Busiest days'

df_ords_prods_merged.loc[(df_ords_prods_merged['orders_day_of_week'] == 6) |
                          (df_ords_prods_merged['orders_day_of_week'] == 2) |
                         (df_ords_prods_merged['orders_day_of_week'] == 5), 'busy_days_of_the_week'] = 'Regular days'

df_ords_prods_merged.loc[(df_ords_prods_merged['orders_day_of_week'] == 3) |
                          (df_ords_prods_merged['orders_day_of_week'] == 4) , 'busy_days_of_the_week'] = 'slowest days'

```

```python
print('-----------------------------------------------------------')

print(color.BLUE + color.BOLD + 'Re-classification of busy days of the week'+ color.END)
print(df_ords_prods_merged['busy_days_of_the_week'].value_counts(dropna=False))
print('-----------------------------------------------------------')
print(color.BLUE + color.BOLD + 'Total number of orders in a day of the week'+ color.END)
print(  df_ords_prods_merged.groupby(['orders_day_of_week','busy_days_of_the_week'])['order_id'].count().sort_values())
```



### Busiest hours of day



```python
df_hod = df_ords_prods_merged['order_hour_of_day'].value_counts()

print(color.BLUE + color.BOLD + 'Total orders in each hour of the day '+ color.END)
print(df_hod)
```

```python
most_orders_threeshold = 2500000
fewest_orders_threeshold = 100000

Most_orders = df_hod[df_hod.gt(most_orders_threeshold, axis=0)]

Average_orders = df_hod[(df_hod.le(most_orders_threeshold, axis=0)) & 
                        df_hod.ge(fewest_orders_threeshold, axis=0)]

Fewest_orders = df_hod[df_hod.lt(fewest_orders_threeshold, axis=0)]


print(color.BLUE + color.BOLD + 'Most orders of the hours:' + color.END, Most_orders.index)

print(color.BLUE + color.BOLD + 'Fewest orders of the hours:' + color.END,Fewest_orders.index)

print(color.BLUE + color.BOLD + 'Average orders of the hours:' + color.END,Average_orders.index)

```

### Create busiest_period_of_day column('Most orders', 'Fewest orders', 'Average orders' )

```python
df_ords_prods_merged.loc[df_ords_prods_merged['order_hour_of_day'].isin(Most_orders.index),
                         'busiest_period_of_day'] = 'Most orders'

df_ords_prods_merged.loc[df_ords_prods_merged['order_hour_of_day'].isin(Fewest_orders.index),
                         'busiest_period_of_day'] = 'Fewest orders'

df_ords_prods_merged.loc[df_ords_prods_merged['order_hour_of_day'].isin(Average_orders.index),
                         'busiest_period_of_day'] = 'Average orders'
```

```python

busiest_hours =  df_ords_prods_merged.groupby(['order_hour_of_day','busiest_period_of_day'])['order_id'].count().sort_values() 
```

```python

print('-----------------------------------------------------------')

print(color.BLUE + color.BOLD + 'Classification of busy hours of the day'+ color.END)

print(df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna=False))

print('-----------------------------------------------------------')
print(color.BLUE + color.BOLD + 'Classification of hours of the day'+ color.END)
print(busiest_hours)
```
