```python
#Import the pandas, NumPy, and os libraries
import pandas as pd
import os


path = r'C:\\Users\\mutta\\Documents\\Project 4\\' 


df_merged_large_pkl = os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl')

orders_products_combined = pd.read_pickle(df_merged_large_pkl)

orders_products_combined.head(10)
```

```python

pd.options.display.float_format = '{:.2f}'.format

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

def validate_mixed_type_check(df):
    mixed_col_count = 0
    for col in df.columns.tolist():
      weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (color.RED + 'Column', col, ' contains mixed type values' + color.END)
        mixed_col_count = mixed_col_count + 1
      else:
        print('Column', col ,'  contains same data type values')

    print(color.BOLD +'### Total number mixed data type column :', mixed_col_count, '###'+ color.END)

```

### Merge data sets

```python



df_orders_products_merged = orders_products_combined.merge(df_prods, on = 'product_id',
                                                           how = 'left', 
                                                           indicator = '_exist')

print('Ordered records left merged on product_id with product records')


```

### Validate Merged data set

```python


print('---------------------------------------------------------')

print(df_orders_products_merged['_exist'].value_counts())

print('---------------------------------------------------------')

print('\n\nNumber of rows in combined order product:', df_orders_products_merged.shape[0])
print('Number of columns in combined order product:', df_orders_products_merged.shape[1])

print(color.RED + 'Total number record count diiference after merge', 
              df_orders_products_merged.shape[0]-orders_products_combined.shape[0],
              ''+ color.END)
    
    
dup_records = orders_products_combined[(orders_products_combined['product_id']==6800) | 
                         (orders_products_combined['product_id']==26520) ]

print(color.BLUE + color.BOLD +
      'Duplicate product ids caused more records after merge, \
        \nTotal number of duplecate product keys:', dup_records.shape[0])
```

```python

print('---------------------------------------------------------')

print(df_orders_products_merged.groupby(pd.Grouper(key="product_id"))['product_id'].count())

     
print('----------------------------------------------')

print('\n##### Mixed data type check ######')

validate_mixed_type_check(df_orders_products_merged)

print('----------------------------------------------')

print('\n##### Missing value check ######')
print( df_orders_products_merged.isnull().sum() )

print('----------------------------------------------')
print('\n##### Duplicate records check ######')
df_orders_products_merged_dups = df_orders_products_merged[df_orders_products_merged.duplicated()]
print(df_orders_products_merged_dups)
print('\n----------------------------------------------')


```
