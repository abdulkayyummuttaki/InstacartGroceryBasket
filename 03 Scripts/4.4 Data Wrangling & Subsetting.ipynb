```python

#Import the pandas, NumPy, and os libraries
import pandas as pd
import os


#Turn your project folder path into a string as outlined in the Exercise text.
path = r'C:\\Users\\mutta\\Documents\\Project 4\\' 

#Import the “orders.csv” data set into your notebook using the os library and 
#the vars_list shortcut, omitting the “eval_set” column in your final dataframe.
vars_list = ['order_id', 'user_id', 'order_number', 'order_dow', 'order_hour_of_day', 'days_since_prior_order']

order_CSV = os.path.join(path, '02 Data', 'Original Data', 'orders.csv')
df_orders = pd.read_csv(order_CSV, index_col = False, usecols = vars_list)

df_orders.round(4)

df_orders.describe()


```

```python
# 2.
# Find another identifier variable in the df_ords dataframe that doesn’t
# need to be included in your analysis as a numeric variable and change it to a suitable format.

df_orders.dtypes

df_orders['order_id'] = df_orders['order_id'].astype('str')
df_orders['user_id'] = df_orders['user_id'].astype('str')
df_orders['order_number'] = df_orders['order_number'].astype('str')

df_orders.dtypes
```

```python
df_orders.describe()
```

```python
#3.
#Look for a variable in your df_ords dataframe with an unintuitive name and
# change its name without overwriting the data frame.

df_orders.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)

df_orders.head(2)
```

```python
#4.
#Your client wants to know what the busiest hour is for placing orders.
#Find the frequency of the corresponding variable and share your findings.

df_groupby = df_orders.groupby(pd.Grouper(key="order_hour_of_day"))['order_id'].count()


print('The busiest hour is for placing orders',df_groupby.idxmax())
```

```python
#5
#Determine the meaning behind a value of 4 in the "department_id" column 
#within the df_prods dataframe using a data dictionary.


dept_CSV = os.path.join(path, '02 Data', '4.4_departments', 'departments.csv')
df_dept = pd.read_csv(dept_CSV, index_col = False)

#Transpose the table
df_dept = df_dept.T

#change the header
new_header = df_dept.iloc[0]
df_dep_t_new = df_dept[1:]
df_dep_t_new.columns = new_header
dept_dict = df_dep_t_new.to_dict('index')

print('Department id 4 in product.csv is:',dept_dict.get('4')['department'])
```

```python
# 6.
#The sales team in your client’s organization wants to know more about breakfast item sales. 
# Create a subset containing only the required information.
product_CSV = os.path.join(path, '02 Data', 'Original Data', 'products.csv')

df_prods = pd.read_csv(product_CSV, index_col = False)

df_breakfast =  df_prods[df_prods['department_id']==14]

df_breakfast
```

```python
dimension = df_breakfast.shape
print('Number of rows in breakfast items:',dimension[0])
print('Number of columns in breakfast items:',dimension[1])
```

```python
#7.
#They’d also like to see details about products that customers might use to throw dinner parties.
# Your task is to find all observations from the entire dataframe that include items 
#from the following departments: alcohol, deli, beverages, and meat/seafood. 
#You’ll need to present this subset to your client.

# function to return key for any value
def get_key(val):
    for key, value in dept_dict.items():
        if val == value:
            return int(key)
 
    return "not found"


df_party_items =  df_prods.loc[df_prods['department_id'].isin([get_key({'department': 'alcohol'}),
                                                                  get_key({'department': 'deli'}),
                                                                  get_key({'department': 'beverages'}),
                                                                  get_key({'department': 'meat seafood'})])]


df_party_items

```

```python
#8.
#It’s important that you keep track of total counts in your dataframes.
# How many rows does the last dataframe you created have?

dimension = df_party_items.shape
print('Number of rows in party items:', dimension[0])
print('Number of columns in party items:', dimension[1])
```
